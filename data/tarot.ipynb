```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
df = pd.read_json('tarot-images.json')
df = pd.json_normalize(df['cards'])
df
```

```python
df['meaning'] = (df['fortune_telling'] + df['meanings.light'] + df['meanings.shadow'])
features = ['name', 'meaning', 'img']
df = df[features]
df
```

```python
df['meaning'] = df['meaning'].apply(lambda x: '. '.join(x))
df
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_meanings(text):
    output = summarizer(text, max_length=30, min_length=10, do_sample=False)
    return output[0]['summary_text']

df['short_meaning'] = df['meaning'].apply(summarize_meanings)
df
```

```python
df.drop('meaning', axis=1, inplace=True)
df.to_csv('tarot_short.csv', index=False)
```
